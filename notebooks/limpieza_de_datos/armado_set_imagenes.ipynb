```python
anonima = 'https://github.com/ianCristianAriel/2024_asistente_compras/blob/main/limpieza_de_datos/anonimaok.csv'
carrefour = 'https://github.com/ianCristianAriel/2024_asistente_compras/blob/main/limpieza_de_datos/preciosCarrefourok.csv'
jumbo= 'https://github.com/ianCristianAriel/2024_asistente_compras/blob/main/limpieza_de_datos/preciosJumbook.csv'
```

```python
import pandas as pd
import requests
from io import BytesIO
from PIL import Image
import os
```

```python
#obtener imagenes

jumbo = pd.read_csv('/content/2024_asistente_compras/limpieza_de_datos/preciosJumbook.csv')

jumbo.head()
```

```python
#obtener imagenes

carrefour = pd.read_csv('/content/2024_asistente_compras/limpieza_de_datos/preciosCarrefourok.csv')

carrefour.head()
```

```python
#obtener imagenes

anonima = pd.read_csv('/content/2024_asistente_compras/limpieza_de_datos/anonimaok.csv')

anonima.head()
```

```python
#anexo los tres DataFrames en uno solo
merged_df = pd.concat([jumbo, carrefour, anonima], ignore_index=True)
```

```python
merged_df.head()
```

```python
urls_column = merged_df.iloc[:, 3]
image_urls = urls_column.tolist()
output_folder = '/content/2024_asistente_compras/DatosPreciosSupermercado/imagenes'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for i, url in enumerate(image_urls):
    response = requests.get(url)
    if response.status_code == 200:
        image = Image.open(BytesIO(response.content))
        image_path = os.path.join(output_folder, f'image_{i}.jpg')
        image.save(image_path)
    else:
        print(f'Error al descargar la imagen {i} desde {url}')

print("Proceso completado.")
```
