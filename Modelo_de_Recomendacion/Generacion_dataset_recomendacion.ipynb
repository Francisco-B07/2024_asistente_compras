```python
import pandas as pd
import random
from datetime import datetime
```

```python
info_super = pd.read_csv('/content/drive/MyDrive/ISPC/3ero/listadoproductoscompletos.csv')
```

```python
info_super.head()
```

```python
# Crear lista de productos
productos_unicos = info_super['Producto'].unique()
categorias = info_super.set_index('Producto')['Categoria'].to_dict()
precios = info_super.set_index('Producto')['Suma de Precio'].to_dict()
palabra_clave = info_super.set_index('Producto')['Primera_palabra'].to_dict()
```

```python
#Se arma dataframe
n_filas = 6000
columnas = ['UserID', 'Product', 'Quantity', 'Date', 'Category', 'Price', 'Palabra_Clave']
data = []
```

```python
#Completar dataset

for _ in range(n_filas):
    user_id = random.randint(1000, 2000)
    product = random.choice(productos_unicos)
    quantity = random.randint(1, 3)
    date = datetime(2023, random.randint(1, 12), random.randint(1, 28))
    category = categorias[product]
    price = precios[product]
    Palabra_Clave = palabra_clave[product]

    data.append([user_id, product, quantity, date, category, price, Palabra_Clave])
```

```python
df = pd.DataFrame(data, columns=columnas)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
df.to_csv('recomendacion_supermercado_dataset.csv', index=False)
```
